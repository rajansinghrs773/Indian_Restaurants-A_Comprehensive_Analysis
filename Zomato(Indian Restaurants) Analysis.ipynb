```python
# Dropping the columns that are not informative purpose

df.drop(["url","address","city_id","latitude","longitude","zipcode","country_id","locality_verbose","currency",
         "opentable_support","takeaway","delivery"],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
df.shape
```

```python
# Renaming the columns name

df.rename(columns={"average_cost_for_two":"2PlatesCost","aggregate_rating":"rating"},inplace=True)
```

```python
df.head(5)
```

```python
# Renaming extra rows to others which have very less frequency

df.name.value_counts().head(10)
```

```python
others_name=df.name.value_counts(ascending=False)
others_name_lessthan_100 = others_name[others_name<100]
```

```python
def name_others(value):
    if (value in others_name_lessthan_100):
        return "others"
    else:
        return value
    
df["name"] = df["name"].apply(name_others)
```

```python
df.name.value_counts()
```

```python
sum(df.city.value_counts()>1000)
```

```python
others_city_name=df.city.value_counts(ascending=False)
others_city_name_lessthan_1000 = others_city_name[others_city_name<1000]
```

```python
def city_others(value):
    if (value in others_city_name_lessthan_1000):
        return "others"
    else:
        return value
    
df["city"] = df["city"].apply(city_others)
```

```python
df.city.value_counts()
```
